```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/dataset/zomato.csv")
df.head()
```

# **Data Cleaning:**

**Deleting redundant columns.**

```python
df.drop("menu_item",axis=1,inplace = True)
df.head()
```

**Renaming the columns.**

```python
df.rename(columns={"name":"rest_name"},inplace=True)
df
```

**Dropping duplicates.**

```python
df.drop_duplicates().head()
df
```

**Cleaning individual columns.**

```python
#missing values of "rest_type" column
miss_values_rest_type = df[df["rest_type"].isnull()].head()

#Filling missing values with "N/A"
df["rest_type"] = df["rest_type"].fillna("N/A")
df.tail()
```

**Remove the NaN values from the dataset**

```python
#fiil rate column with "N/A"
df["rate"] = df["rate"].fillna("N/A")

#fill phone with "N/A"
df["phone"] = df["phone"].fillna("N/A")

#fill dish_liked with "N/A"
df["dish_liked"] = df["dish_liked"].fillna("N/A")

#fill approx_cost(for two people) with "N/A"
df["approx_cost(for two people)"] = df["approx_cost(for two people)"].fillna("N/A")

#fill cuisines column with "N/A"
df["cuisines"] = df["cuisines"].fillna("N/A")

# fill rest_type missing values with "N/A"
df["rest_type"] = df["rest_type"].fillna("N/A")

##Remove NAN values from "location"
df = df.dropna(subset=["location"])
```

**Check for some more Transformations**

```python
df.rename(columns={"phone":"Phone_Number"},inplace=True)
df.head()
```

# **Data Visualization:**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

**Restaurants delivering Online or not**

```python
sns.set_style("whitegrid")

# Create figure
fig,ax = plt.subplots(figsize=(7, 5))

## Countplot
sns.countplot(x="online_order",data= df,hue="online_order",edgecolor = "black")

## Title and labels
ax.set_title("Restaurants Offering Online Delivery or Not", fontsize=14, fontweight="bold")
ax.set_xlabel("Online Delivery Available?", fontsize=10)
ax.set_ylabel("Number of Restaurants", fontsize=10)
```

**Restaurants allowing table booking or not**

```python
sns.set_style("whitegrid")
# Create figure
fig,ax = plt.subplots(figsize=(7, 5))

# Countplot
sns.countplot(x="book_table",data= df,hue="book_table",palette="Set2",edgecolor="black")

## Add title and labels
ax.set_title("Restaurants Allowing Table Booking or Not", fontsize=14, fontweight="bold", color="#333")
ax.set_xlabel("Table Booking Available?", fontsize=10)
ax.set_ylabel("Number of Restaurants", fontsize=10)
```

**Table booking Rate vs Rate**

```python
#Create figure
plt.figure(figsize=(8,6))

## Barplot
sns.barplot(x="book_table",y="rate",data= df,hue="book_table",palette="crest")

## Add title and labels
plt.title("Average Rating vs Table Booking Availability", fontsize=16, fontweight="bold", color="#333")
plt.xlabel("Table Booking Available?", fontsize=13)
plt.ylabel("Average Rating", fontsize=13)
```

**Best Location**

```python
## Top 10 locations by total votes
bl = df.groupby("location")["votes"].sum().sort_values(ascending=False).head(10)

# Create figure
fig, ax = plt.subplots(figsize=(16, 6))

## Barplot
sns.barplot(x=bl.index,y=bl.values,palette="viridis",edgecolor="black")

## Title and labels
ax.set_title("Top 10 Locations by Total Votes", fontsize=18, fontweight="bold")
ax.set_xlabel("Locations" ,fontsize=14)
ax.set_ylabel("Total Votes", fontsize=14)
ax.set_xticklabels(ax.get_xticklabels(), rotation=20, ha='right')
```

**Relation between Location and Rating**

```python
import numpy as np

#Cleaned RATE column
df['rate'] = df['rate'].replace('N/A', np.nan)
df['rate'] = df['rate'].astype(str).str.replace('/5', '', regex=False)
df['rate'] = pd.to_numeric(df['rate'], errors='coerce')

## Calculate average rating by location
rel = df.groupby("location")["rate"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(15, 7))  ## Create figure

#Barplot with Palette
sns.barplot(y=rel.index,x=rel.values,palette="viridis",edgecolor ="black")

## Titles and labels
plt.title("Top 10 Locations by Average Restaurant Rating", fontsize=18, fontweight="bold")
plt.xlabel("Average Rating", fontsize=14)
plt.ylabel("Location",fontsize=14)
```

**Restaurant Type**

```python
# Top 15 restaurant types
t =  df["rest_type"].value_counts().head(15)

## Create figure
fig,ax = plt.subplots(figsize=(18,6))

## Barplot with palette
sns.barplot(y=t.values,x=t.index,palette="viridis")

## Titles and labels
ax.set_xlabel("Restaurant Type", fontsize=14)
ax.set_ylabel("Count",fontsize=14)
ax.set_title("Restaurant Type",fontsize =18, fontweight="bold")
ax.set_xticklabels(ax.get_xticklabels(),rotation = 20,ha= "right")
```

**Gaussian Rest type and Rating**

```python
##Cleaned RATE column
df["rate"] = df["rate"].replace("N/A",np.nan)
df["rate"] = df["rate"].astype(str).str.replace("/5","",regex=False)
df["rate"] = pd.to_numeric(df["rate"],errors = "coerce")

# Calculated top 15 rest_types by average rating
t =  df.groupby("rest_type")["rate"].mean().sort_values(ascending=False).head(15)

# Created the figure
fig,ax = plt.subplots(figsize=(14,8))
sns.barplot(y=t.index,x=t.values,palette="viridis",edgecolor="black")

# Added title and labels
ax.set_title("Top 15 Rest_type By Average Rating",fontsize = 14,fontweight="bold")
ax.set_xlabel("Avg. Rating")
ax.set_ylabel("Restaurant Type")
```

**Types of Services**

**Relation between Type and Rating**

```python
##Cleaned RATE column
df["rate"] = df["rate"].replace("N/A",np.nan)
df["rate"] = df["rate"].astype(str).str.replace("/5","",regex=False)
df["rate"] = pd.to_numeric(df["rate"],errors = "coerce")

sns.set_style("whitegrid")
fig,ax = plt.subplots(figsize=(16,6))

# Violin plot with better aesthetics
sns.violinplot(x="listed_in(type)",y="rate",data=df,palette="Set2")

# Add title and labels
ax.set_title("Distribution of Ratings Across Service Types",fontsize=16,fontweight="bold")
ax.set_xlabel("Service Type")
ax.set_ylabel("Rating")
```

**Cost of Restaurant**

**No. of restaurants in a Location**

```python
## Prepared data
rl = df.groupby("location")["rest_name"].count().sort_values(ascending=False).head(20)

sns.set_style("whitegrid")
fig,ax = plt.subplots(figsize=(16,6))

## Created bar plot with color palette
sns.barplot(x=rl.index,y=rl.values,palette="mako",edgecolor="black")
ax.set_xticklabels(ax.get_xticklabels(),rotation=60,ha="right")

## Title and labels
ax.set_title("Top 20 Locations with Most Restaurants", fontsize=18, fontweight='bold')
ax.set_xlabel("Location", fontsize=14)
ax.set_ylabel("Number of Restaurants", fontsize=14)
```
